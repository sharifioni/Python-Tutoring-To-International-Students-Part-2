# Nested For loops
* Using a for loop inside another for loop

```python
count=0
for i in range(1,11):
    for j in range(1,5):
        print("i=",i,"\tj=",j)
        count=count+1
print(count)
        
```

```python
lower_range=int(input("enter a number: "))
upper_range=int(input("enter a number: "))
if lower_range<upper_range:
    for i in range(lower_range,upper_range+1):
        for j in range(1,11):
            print(f"{i:<5} {'*':<5} {j:<5} {'=':<5} {i*j:<5}")
            
        print("------------------------------------")
        
        
else:
    for i in range(lower_range,upper_range+1,-1):
        for j in range(1,11):
            print(f"{i:<5} {'*':<5} {j:<5} {'=':<5} {i*j:<5}")
            
        print("------------------------------------")
```

# while loops
* While loop runs based on a condition. If the condition is True, it will keep looping, and when condition becomes it'll stop

* while condition:
    statement1.....
    statement2.....
    ...............

```python
x=10

while x<20:
    print(x)
    x=x+1
```

```python
for i in range(1,11):
    print(i)
```
