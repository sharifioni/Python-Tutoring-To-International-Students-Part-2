```python
#You have to print out even numbers between two ranges (lower range and upper range) using nested for loop (C++, Java)  
lower=int(input("Enter lower range")) #5
upper=int(input("Enter upper range")) #26


if lower < upper:
    for i in range(lower,upper+1):  #5 , 27
        if i%2==0:
            print("This number is even",i)
else:
    print("Sorry the lower number is higher than the upper number")
#While
        
```
