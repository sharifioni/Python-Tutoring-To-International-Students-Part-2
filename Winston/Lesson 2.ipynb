```python
#Exercise 1
def factorial(n: int) -> int: 
    """Return n! for positive values of n. 
     >>> factorial(1) 
     1 
     >>> factorial(2) 
     2 
     >>> factorial(3) 
     6 
     >>> factorial(4) 
     24 
     """ 
    fact = 1 
    for i in range(2,n+1): 
        fact = fact * n 
 
    return fact


test_passed=0
test_failed=0


#Testing for factorial(1)
a=factorial(1)
print("Testing factorial(1)")
print("Expected Result: 1","Actual Result:",a)
if a==1:
    print("Test Passed")
    test_passed+=1
else:
    print("Test failed")
    test_failed+=1

#Testing for factorial(1)
a=factorial(2)
print("Testing factorial(2)")
print("Expected Result: 2","Actual Result:",a)
if a==2:
    print("Test Passed")
    test_passed+=1
else:
    print("Test failed")
    test_failed+=1

#Testing for factorial(3)
a=factorial(3)
print("Testing factorial(3)")
print("Expected Result: 6","Actual Result:",a)
if a==6:
    print("Test Passed")
    test_passed+=1
else:
    print("Test failed")
    test_failed+=1

#Testing for factorial(4)
a=factorial(4)
print("Testing factorial(4)")
print("Expected Result: 24","Actual Result:",a)
if a==24:
    print("Test Passed")
    test_passed+=1
else:
    print("Test failed")
    test_failed+=1
    
print(test_passed,"tests passed for Exercise 1")
print(test_failed,"tests failed for Exercise 1")
```

```python
#Exercise 2
def triangle_type(angle1:int,angle2:int,angle3:int):
    
    """Returning the type of triangle
    >>>triangle_type(60,60,60)
    Acute triangle
    
    >>>triangle_type(90,30,60)
    Right triangle
    
    >>>triangle_type(100,50,30)
    Obtuse triangle
    
    >>>triangle_type(70,60,60)
    This is not a triangle
    """
    
    if angle1+angle2+angle3>180 or angle1+angle2+angle3>180:
        return "This is not a triangle"
        
        
    elif angle1 < 90  and angle2<90 and angle3<90:
        return "Acute triangle"
    
    elif angle1==90 or angle2==90 or angle3==90:
        return "Right triangle"
    
    elif angle1 > 90 or angle2>90 or angle3>90:
        return "Obtuse triangle"
    

test_passed=0
test_failed=0

#Testing for triangle_type(60,60,60)
a=triangle_type(60,60,60)
print("Testing triangle_type(60,60,60)")
print("Expected Result: Acute triangle","Actual Result:",a)
if a=="Acute triangle":
    print("Test Passed")
    test_passed+=1
else:
    print("Test failed")
    test_failed+=1
    
    
#Testing for triangle_type(90,30,60)
a=triangle_type(90,30,60)
print("Testing triangle_type(90,30,60)")
print("Expected Result: Right triangle","Actual Result:",a)
if a=="Right triangle":
    print("Test Passed")
    test_passed+=1
else:
    print("Test failed")
    test_failed+=1
    
    
#Testing for triangle_type(100,50,30)
a=triangle_type(100,50,30)
print("Testing triangle_type(100,50,60)")
print("Expected Result: Obtuse triangle","Actual Result:",a)
if a=="Obtuse triangle":
    print("Test Passed")
    test_passed+=1
else:
    print("Test failed")
    test_failed+=1
    
#Testing for triangle_type(70,60,60)
a=triangle_type(70,60,60)
print("Testing triangle_type(70,60,60)")
print("Expected Result: This is not a triangle","Actual Result:",a)
if a=="This is not a triangle":
    print("Test Passed")
    test_passed+=1
else:
    print("Test failed")
    test_failed+=1

print(test_passed,"tests passed for Exercise 2")
print(test_failed,"tests failed for Exercise 2")
```

```python
#Exercise 3
def discount(age:int,product_price:int):
    """Function to calculate discount according to customer's age
    >>>discount(85,100)
    20
    
    >>>discount(70,150)
    24
    
    >>>discount(55,100)
    13
    
    >>>discount(25,230)
    11.5
    
    >>>discount(15,170)
    0
    """
    if age>=80:
        discount=product_price*0.2
        return discount
    
    elif age>=60:
        discount=product_price*0.16
        return discount
    
    elif age>=40:
        discount=product_price*0.13
        return discount
    
    elif age>=18:
        discount=product_price*0.05
        return discount
    
    elif age<18:
        discount=0
        return discount
    
test_passed=0
test_failed=0

#Testing for discount(85,100)
a=discount(85,100)
print("Testing discount(85,100)")
print("Expected Result: 20","Actual Result:",a)
if a==20:
    print("Test Passed")
    test_passed+=1
else:
    print("Test failed")
    test_failed+=1
    
#Testing for discount(70,150)
a=discount(70,150)
print("Testing discount(70,150)")
print("Expected Result: 24","Actual Result:",a)
if a==24:
    print("Test Passed")
    test_passed+=1
else:
    print("Test failed")
    test_failed+=1
    

#Testing for discount(55,100)
a=discount(55,100)
print("Testing discount(55,100)")
print("Expected Result: 13","Actual Result:",a)
if a==13:
    print("Test Passed")
    test_passed+=1
else:
    print("Test failed")
    test_failed+=1
    
    
#Testing for discount(25,230)
a=discount(25,230)
print("Testing discount(25,230)")
print("Expected Result: 11.5","Actual Result:",a)
if a==11.5:
    print("Test Passed")
    test_passed+=1
else:
    print("Test failed")
    test_failed+=1
    
#Testing for discount(15,100)
a=discount(15,170)
print("Testing discount(15,170)")
print("Expected Result: 0","Actual Result:",a)
if a==0:
    print("Test Passed")
    test_passed+=1
else:
    print("Test failed")
    test_failed+=1
    
print(test_passed,"tests passed for Exercise 3")
print(test_failed,"tests failed for Exercise 3")
```
