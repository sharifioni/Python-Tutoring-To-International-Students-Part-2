# for loops
* used to loop over an iterable
* an iterable can be string, list, tuple, dictionary, range, set, zip, dictionary.keys(), dictionary.values()
* for loop runs from the first item until last item from an iterable 
* everytime a for loop runs, we call it as an "iteration"
* for (variable name) in iterable:
      statement1.........
      statement2.......
      .................

```python
#for loop over a iterable (string)
for i in "Programming":
    print(i)
    
```

```python
# Take a string from the user and count the number of vowels in that string
sentence = input("Enter any sentence: ")
count = 0
for i in sentence:
    if i=="a" or i=="e" or i=="i" or i=="o" or i=="u":
        count = count+1

print(count)
        
        
        
```

# Range function with for loops

* range()
* it accepts atmost 3 integer values / arguments
* start value, stop value, step value

```python
# range function with one argument with for loop
# by default start value is 0
# stop value is the value that you have given
# step value is 1
for a in range(10):
    print(a)
```

```python
# range function with two arguments with for loop
#  start value is first argument
# stop value is the second argument
# step value is by default 1
a=6
for a in range(5,100):
    print(a)
```

```python
# range function with three arguments with for loop
#  start value is first argument
# stop value is the second argument
# step value is the third argument

for a in range(10,100,3):
    print(a)
```

```python

for a in range(99,0,-1):
    print(a)
```

```python
number=int(input("Enter any number: "))
for i in range(1,21):
    print(f"{number}*{i}={number*i}")
```
