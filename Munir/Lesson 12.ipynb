```python
# For taking care of spaces 
k=0

# For controlling number of rows
for i in range(10):
    print(" "*k,end="")
    #number of aestriks
    for j in range(10-i):
        print("*",end="")
    k+=1
    print()
```

```python
# For taking care of spaces 
k=9

# For controlling number of rows
for i in range(10,0,-1):
    print(" "*k,end="")
    #number of aestriks
    for j in range(10-(i-1)):
        print("*",end="")
    k-=1
    print()
```

```python
def farenheit(Cel):
    F = (9 / 5) * Cel + 32
    return F

for celsius in range(0,101):
    print(farenheit(celsius))
    
    
#Celsius= 0  , Farenheit equivalent=32.0
#Celsius= 1  , Farenheit equivalent=33.8
    
```

```python
def farenheit(Cel):
    F = (9 / 5) * Cel + 32
    return F

for celsius in range(0,101):
    print(f"Celsius= {celsius:<10} Farenheit equivalent = {farenheit(celsius): > 10.2f}")
    
    
#Celsius= 0  , Farenheit equivalent=32.0
#Celsius= 1  , Farenheit equivalent=33.8
```
