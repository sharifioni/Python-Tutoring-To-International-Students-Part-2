```python
#List
grades=[55,66,77,88,95]

for score in grades:  #score=66
    if score<50:
        print("You got F and your score was",score)
        
    elif score>=50 and score <=59:
        print("You got D and your score was",score)
        
    elif score>=60 and score <=69:
        print("You got C and your score was",score)
        
    elif score>=70 and score <=79:
        print("You got B and your score was",score)
        
    elif score>=80 and score <=89:
        print("You got A and your score was",score)
    
    elif score>=90 and score <=100:
        print("You got A+ and your score was",score)
```

# Nested for loops

> A for loop inside another for loop


```python
for i in range(1,21):
    for j in range(1,10):
        print(i,j,end="\t")
        
    print("\n")
```

```python
num1=int(input("Enter lower number: "))#5
num2=int(input("Enter higher number: ")) #15

for i in range(num1,num2+1):
    for j in range(1,11):
        print(f"{i}\t*\t{j}\t=\t{i*j}")
    print("\n")
        
# Put a newline after every table of i
# Have a look at while loops
```
