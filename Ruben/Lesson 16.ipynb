Temperature Conversion) Implement a fahrenheit function that returns the
Fahrenheit equivalent of a Celsius temperature. Use the following formula:
F = (9 / 5) * C + 32
Use this function to print a chart showing the Fahrenheit equivalents of all Celsius temperatures in the range 0â€“100 degrees. Use one digit of precision for the results. Print the

```python
#You have to create a farenheit function
#It'll convert celsius temperature to farenheit
#It'll return temperature in farenheit
#Print all farenheit equivalents of celsius temperatures from 0 to 100 degrees

def farenheit(cel):
    F = (9 / 5) * cel + 32
    print("Celcsius=",cel,"Farenheit=",F)
    
for celsius in range(0,101):
    farenheit(celsius)
    
```
